```python
# converting the date from DD/MM/YYYY to YYYY-MM-DD
data_info = pd.to_datetime(
    air_quality_data['Date'],
    format='%Y-%m-%d')
print(data_info)
```

```python
# 1) Replace NaN with empty strings, then convert to string dtype
time_info = time_info.fillna('').astype(str)

# 2) Swap '.' with ':' only for non-empty strings
time_info = time_info.apply(lambda x: x.replace('.', ':') if x else x)

# 3) Parse as times (HH:MM:SS), coerce invalid to NaT, then extract time
time_info = pd.to_datetime(time_info, format='%H:%M:%S', errors='coerce').dt.time

print(time_info)
```

```python
from datetime import time

# Convert "Time" strings like "18.00.00" to datetime objects
# using the format H.M.S, then extract just the time.
time_info = pd.to_datetime(
    air_quality_data['Time'],
    format='%H.%M.%S',
    errors='coerce'
).dt.time

print(time_info)
```

```python
print(type(data_info))
print(type(time_info))
```

```python
# combining 2 series to a pandas dataframe
data_time = pd.concat([data_info, time_info], axis=1)
```
